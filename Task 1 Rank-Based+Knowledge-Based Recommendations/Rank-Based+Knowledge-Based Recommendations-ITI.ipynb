#### 1. How To Find The Most Popular Movies

For this notebook, we have a single task. The task is that no matter the user, we need to provide a list of the recommendations based on simply the most popular items.

For this task, we will consider what is "most popular" based on the following criteria:

- A movie with the highest average rating is considered best
- With ties, movies that have more ratings are better
- A movie must have a minimum of 5 ratings to be considered among the best movies
- If movies are tied in their average rating and number of ratings, the ranking is determined by the movie that is the most recent rating

With these criteria, the goal for this notebook is to take a **user_id** and provide back the **n_top** recommendations. Use the function below as the scaffolding that will be used for all the future recommendations as well.

````python
def popular_recommendations(user_id=0, n_top=10):
    '''
    INPUT:
    user_id - the user_id of the individual you are making recommendations for
    n_top - an integer of the number recommendations you want back
    OUTPUT:
    top_movies - a list of the n_top recommended movies by movie title in order best to worst
    '''
    pop_movies = pd.DataFrame(reviews.groupby('movie_id')[
                              'rating'].mean().sort_values(ascending=False)).reset_index()
    movie_id_freq_series = reviews['movie_id'].value_counts()
    rating_freq = pd.DataFrame(
        {'movie_id': movie_id_freq_series.index, 'rating_freq': movie_id_freq_series.values})
    pop_frq_movies = pd.merge(pop_movies, rating_freq,
                              on='movie_id', how='outer')

    # 3)A movie must have a minimum of 5 ratings to be considered among the best movies
    pop_frq_movies = pop_frq_movies[pop_frq_movies['rating_freq'] >= 5]

    # 4)If movies are tied in their average rating and number of ratings, the ranking is determined by the movie that is the most recent rating

    reviews['recent_in_days'] = (
        datetime.now() - reviews['date']).apply(lambda x: x.days)
    reviews.sort_values(by='recent_in_days')
    rating_recent = pd.DataFrame(reviews.groupby(
        'movie_id')['recent_in_days'].min().sort_values()).reset_index()
    best_movies_df = pd.merge(pop_frq_movies, rating_recent, on='movie_id')

    final_best_df = best_movies_df.sort_values(
        by=['rating', 'rating_freq', 'recent_in_days'], ascending=[False, False, True])
    movies_df = movies[['movie_id', 'movie', 'genre', 'date']]
    top_movies = pd.merge(final_best_df, movies_df)

    return top_movies.head(n_top)  # a list of the n_top movies as recommended
````

````python
popular_recommendations(n_top=20)
````
